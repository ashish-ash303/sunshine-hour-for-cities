# Sunshine Hour For Cities Across the World

The "Worldwide Sunshine Hours Dataset" is a comprehensive collection of data documenting the average annual sunshine hours for cities across the globe. This dataset provides valuable information about the amount of sunlight received in different cities, aiding in various fields such as climate research, solar energy planning, tourism, and agriculture.


- Dataset Url: https://www.kaggle.com/datasets/bilalwaseer/sunshine-hours-for-cities-around-the-world

### Objectives:

- Sort the Cities Based on Sunshine recieved in month Jan and plot the first 20 Country.
    
- Sort the City Based on the Sunshine recieved and plot the first 20 Country who recieved the most sunshine during the whole 
      year
    
- Which Country/City has recieved the highest sunshine in June month plot them.
  
- Analysis the Data for India Country and Cities.
    
- Draw the Analysis for Australia Country.

## Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Downloading the dataset

```python
import opendatasets as od

dataset_url="https://www.kaggle.com/datasets/bilalwaseer/sunshine-hours-for-cities-around-the-world"

od.download(dataset_url)
```

```python
dataset_dir='./sunshine-hours-for-cities-around-the-world/Sunshine hours for cities in the world.csv'
```

### Importing the Dataset

## Exploratory Data Analysis for India

```python
India_df=df.query('Country == "India"')
(India_df)
```

```python
India_df.plot.bar(x='City',y='Year')
```

```python
plt.figure(figsize=(16,12))
sns.barplot(data=India_df,x="City",y="Jan",palette = "Oranges")
plt.title("Sunshine Recieved by Different Cities of India in Januaray Month",fontsize=20,pad=15)
plt.xlabel('Cities',fontsize=17)
plt.ylabel('Month January',fontsize=17)


plt.tick_params(axis='x',labelsize=10,which='major')
plt.tick_params(axis='y',labelsize=10,which='major')
```

- Mumbai and Banglore are recieving more sunshine in january month in India.It is true because in january Winters are in peak that why Delhi and kolkata received less sunshine compared to other three cities of india.

```python
fig, ax = plt.subplots(figsize =(16, 9))
 
# Horizontal Bar Plot
ax.barh(India_df.City, India_df.Sep,color='skyblue')

plt.title("Sunshine Recieved by Different Cities of India In September Month",fontsize=20,pad=15)
plt.xlabel('Cities',fontsize=17)
plt.ylabel('Month September',fontsize=17)


plt.tick_params(axis='x',labelsize=15,which='major')
plt.tick_params(axis='y',labelsize=15,which='major')

```

- In Month September most sunshine is recieved by `Delhi` as it's only of the hottest city of india during summer.

```python
plt.figure(figsize=(8,6))

# declaring exploding pie
explode = [0.01, 0.03, 0.01, 0.02, 0.04]
# define Seaborn color palette to use
palette_color = sns.color_palette('dark')
  
# plotting data on chart
plt.pie(India_df.Year, labels=India_df.City, colors=palette_color,
        explode=explode, autopct='%.0f%%')
  
# displaying chart
plt.show()
```

- Chennai Followed by Delhi got the highest Sunshine throughout the Whole Year.

## Sort the Cities Based on Sunshine recieved and plot the first 20 Country

```python
plt.figure(figsize=(12,8))
sns.histplot(df['Year'],kde=True,color='blue',bins=12)
plt.title("Year Values and their count",fontsize=15,pad=20)

plt.xlabel('Year_Values',fontsize=12)
plt.ylabel('Count',fontsize=12)

plt.tick_params(axis='x',which='major',labelsize=8)
plt.tick_params(axis='y',which='major',labelsize=8)
```

- Maximum Count of values are lies between 2000 to 3000 

```python
sort_df=df.sort_values(by='Year',ascending=False)[:20]
sort_df
```

```python
fig, ax = plt.subplots(figsize =(16, 9))
 

ax.bar(sort_df.City,sort_df.Year,color='skyblue')
plt.title("Top 20 Cities who recieved maximum sunshine Throughout the Year",fontsize=20,pad=15)
plt.xlabel('Cities',fontsize=17)
plt.ylabel('Year Values',fontsize=17)


plt.tick_params(axis='x',labelsize=15,which='major',rotation=90)
plt.tick_params(axis='y',labelsize=15,which='major')
```

```python
plt.figure(figsize=(16,12))
sns.barplot(data=sort_df,x="City",y="Year",palette = "Blues")
plt.title("Top 20 Countries Which gets Most Sunshine throughout the Year.",fontsize=12)
plt.xlabel('Cities',fontsize=17)
plt.ylabel('Year',fontsize=17)


plt.tick_params(axis='x',labelsize=15,which='major',rotation=90)
plt.tick_params(axis='y',labelsize=15,which='major')
```

- Yuma Recives the highest Sunshine During the Whole Year

```python
plt.figure(figsize=(15,12))
sns.histplot(sort_df['Country'],color='blue',bins=11)

plt.title("Countries and thier Count in top 20 Cities which recieved most sunshine in whole year.",fontsize=12,pad=15)
plt.xlabel('Countries',fontsize=17)
plt.ylabel('Count',fontsize=17)

plt.tick_params(axis='x',labelsize=15,which='major',rotation=90)
plt.tick_params(axis='y',labelsize=15,which='major')
plt.grid(linewidth=0.4)
```

- USA has highest 7 Cities Which recieves highest sunshine during the Whole year.

## Which Country/City has recieved the highest sunshine in June Month.

```python
june_df=df.sort_values(by='Jun',ascending=False)[:20]
june_df
```

```python
plt.figure(figsize=(12,8))
sns.histplot(june_df.Country,color='orange',bins=8)

plt.title("Countries and thier Count in top 20 Cities which recieved most sunshine in June Month.",fontsize=12,pad=15)
plt.xlabel('Countries',fontsize=17)
plt.ylabel('Count',fontsize=17)

plt.tick_params(axis='x',labelsize=15,which='major',rotation=90)
plt.tick_params(axis='y',labelsize=15,which='major')
plt.grid(linewidth=0.4)
```

- In June month USA has recieved highest Sunshine With 10 Cities are in Top 20 for getting most sunshine. It means in month of June it is mostly Summer in USA.

```python
Aus_df=df.query('Country == "Australia"')
(Aus_df)
```

```python
plt.figure(figsize=(12,8))

explode = [0.01, 0.01, 0.01, 0.01, 0.01, 0.01, 0.01, 0.01, 0.01, 0.01, 0.01, 0.01, 0.01, 0.01, 0.01]
palette_color = sns.color_palette('dark')
plt.pie(Aus_df.Year, labels=Aus_df.City, colors=palette_color,
         explode=explode,autopct='%.0f%%')
plt.show()
```
